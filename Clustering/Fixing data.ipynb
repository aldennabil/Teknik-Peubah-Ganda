```python
import pandas as pd
df = pd.read_csv('Data Produk long.csv', sep =";")
df
```

```python
# Format NamaProv agar sesuai dengan GADM dan laporkan hasil pencocokan
gadm_provinsi = [
    'Aceh', 'Bali', 'Bangka Belitung', 'Banten', 'Bengkulu', 'Gorontalo',
    'Jakarta Raya', 'Jambi', 'Jawa Barat', 'Jawa Tengah', 'Jawa Timur',
    'Kalimantan Barat', 'Kalimantan Selatan', 'Kalimantan Tengah',
    'Kalimantan Timur', 'Kalimantan Utara', 'Kepulauan Riau', 'Lampung', 'Maluku',
    'Maluku Utara', 'Nusa Tenggara Barat', 'Nusa Tenggara Timur', 'Papua',
    'Papua Barat', 'Riau', 'Sulawesi Barat', 'Sulawesi Selatan',
    'Sulawesi Tengah', 'Sulawesi Tenggara', 'Sulawesi Utara', 'Sumatera Barat',
    'Sumatera Selatan', 'Sumatera Utara', 'Yogyakarta'
 ]

# Normalisasi NamaProv di df agar sesuai dengan GADM (misal: strip, title-case, dll)
df['NamaProv_norm'] = df['NamaProv'].str.strip().str.title()

# Cek kecocokan NamaProv_norm dengan GADM
laporan = []
for prov in df['NamaProv_norm'].unique():
    if prov in gadm_provinsi:
        laporan.append(f"✔️ {prov} ditemukan di GADM")
    else:
        laporan.append(f"❌ {prov} TIDAK ditemukan di GADM")

# Tampilkan laporan
for l in laporan:
    print(l)
```

```python
# Perbaiki NamaProv yang tidak sesuai dan cek provinsi GADM yang tidak termap
replace_dict = {
    'Kepulauan Bangka Belitung': 'Bangka Belitung',
    'Di Yogyakarta': 'Yogyakarta',
    'Dki Jakarta': 'Jakarta Raya',
}
df['NamaProv_norm'] = df['NamaProv_norm'].replace(replace_dict)

# Isi NA dengan Kalimantan Utara
df['NamaProv_norm'] = df['NamaProv_norm'].fillna('Kalimantan Utara')

# Cek ulang kecocokan NamaProv_norm dengan GADM
laporan = []
provinsi_termap = set()
for prov in df['NamaProv_norm'].unique():
    if prov in gadm_provinsi:
        laporan.append(f"✔️ {prov} ditemukan di GADM")
        provinsi_termap.add(prov)
    else:
        laporan.append(f"❌ {prov} TIDAK ditemukan di GADM")

# Cari provinsi di GADM yang tidak termap di data
gadm_tidak_termap = set(gadm_provinsi) - provinsi_termap

# Tampilkan laporan pencocokan
for l in laporan:
    print(l)

print("\nProvinsi di GADM yang tidak termap di data:")
for prov in gadm_tidak_termap:
    print(f"❌ {prov} tidak termap di data")
```

```python
# Cek apakah masih ada row yang belum termap di GADM
belum_termapped = df[~df['NamaProv_norm'].isin(gadm_provinsi)]
if belum_termapped.shape[0] > 0:
    print(f"Masih ada {belum_termapped.shape[0]} baris yang belum termap di GADM.")
    display(belum_termapped[['NamaProv', 'NamaProv_norm']].drop_duplicates())
else:
    print("Semua baris sudah termap ke GADM.")
```

```python

import geopandas as gpd

# Load GADM data
gadm_path = r"c:\Alden\Kuliah\Semester 5\Teknik-Peubah-Ganda\Tugas akhir\gadm41\gadm41_IDN_2.dbf"
gadm = gpd.read_file(gadm_path)

# Filter GADM to only include level kota/kabupaten
gadm_kabkot = gadm[gadm['ENGTYPE_2'].isin(['Regency', 'City'])]

# Normalize names in GADM (title-case and strip)
gadm_kabkot['KabKot_norm'] = gadm_kabkot['NAME_2'].str.strip().str.title()

# Normalize names in df (remove "Kab." and "Kota ", then title-case and strip)
df['KabKot_norm'] = df['KabKot'].str.replace(r"^Kab\. ", "", regex=True).str.replace(r"^Kota ", "", regex=True).str.strip().str.title()

# Perform mapping
mapped_kabkot = df['KabKot_norm'].isin(gadm_kabkot['KabKot_norm'])

# Count mapped and unmapped rows
mapped_count = mapped_kabkot.sum()
unmapped_count = (~mapped_kabkot).sum()

# Get list of unmapped kabupaten/kota
unmapped_kabkot = df.loc[~mapped_kabkot, 'KabKot'].unique()

# Print results
print(f"Jumlah yang sudah termap: {mapped_count}")
print(f"Jumlah yang belum termap: {unmapped_count}")
print("Kabupaten/Kota yang belum termap:")
print(unmapped_kabkot)
```
